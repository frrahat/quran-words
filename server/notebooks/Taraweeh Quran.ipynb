```python
import sqlite3
db_arabic_con = sqlite3.connect('../databases/quran_arabic.db')

non_waqfs = {
    '\u06df',
    '\u06e2',
    '\u06e5',
    '\u06e6',
}
def is_waqf_char(char):
    if char in non_waqfs:
        return False
    return (char >= '\u0610' and char <= '\u0615') or (char >= '\u06D6' and char <= '\u06ED')


def get_query_by_day(day_index, selections, table_name, order_by="sura, ayah"):
    day_verse = day_verses[day_index]
    start_sura_ayah, _ = day_verse[0].split(' - ')
    _, end_sura_ayah = day_verse[-1].split(' - ')
    
    start_sura, start_ayah = start_sura_ayah.split(":")
    end_sura, end_ayah = end_sura_ayah.split(":")
    
    selection_text = f"Select {selections} from {table_name}"
    
    union_queries = [
        f"{selection_text} where sura = {start_sura} and ayah >= {start_ayah}",
        f"{selection_text} where sura = {end_sura} and ayah <= {end_ayah}",
    ]
    
    for sura in range(int(start_sura) + 1, int(end_sura)):
        union_queries.insert(-1, f"{selection_text} where sura = {sura}")
    
    return " union ".join(union_queries) + f" order by {order_by}"

def get_verses_query_by_day(day_index):
    return get_query_by_day(day_index, "sura, ayah, text", "verses")


def get_verses_by_day(day_index):
    query = get_verses_query_by_day(day_index)
    cursor = db_arabic_con.cursor()
    data = cursor.execute(query).fetchall()
    cursor.close()
    return data

def get_last_segment(text):
    has_last_segment = False
    for r_index, char in enumerate(text[::-1]):
        if is_waqf_char(char):
            has_last_segment = True
            break
    offset = 1 if has_last_segment else 0
    return text[len(text)-r_index-1 + offset:]

def get_last_segments_by_day(day_index):
    verses = get_verses_by_day(day_index)    
    return [(sura, ayah, get_last_segment(text)) for sura, ayah, text in verses]

print(get_last_segments_by_day(3))
```
